```
# #Definindo as variáveis dependentes e independentes
x = df_full.iloc[:,1:13] #Caso sejam removidos dados nao numericos
y = df_full.iloc[:,13]
```

```
import numpy as np
#Convertendo para array os dados de treinamento
x = np.asarray(x).astype('float32')
y = np.asarray(y).astype('float32')
len(df) - len(x)
```

```
#Separando os dados em dados de treinamento e teste
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 42)
```

```
# Importando as bibliotecas do pacote Keras paratrabalhar com RNAs
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```
#Inicialização da rede
classifier = Sequential()
classifier
```

### Inicializando a RNA

```
#Adicionando a primeira Camada da rede
classifier.add(Dense(units = 7, kernel_initializer='uniform', activation = 'relu', input_dim = 12))

#Adicionando a segunda camada da rede
classifier.add(Dense(units = 7, kernel_initializer='uniform', activation = 'relu'))

#Adicionando a camada de saída
classifier.add(Dense(units = 1, kernel_initializer='uniform', activation='sigmoid'))

# Compilando a RNA
classifier.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```
#Adequando a RNA ao conjunto de exemplos
classifier.fit(x_train, y_train, batch_size = 12, epochs = 5)
#batch_size: o número de entradas processadas simultaneamente
#epoch indica a passagem completa sobre todo o dataset
```

```
# Fazendo as previsões sobre o conjunto de testes
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)

# Construindo a Matriz de Confusão
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred) 
```

## Comparando com Árvores de decisão (trabalho anterior)

Na implementação de árvores de decisão para solução deste mesmo problema, também foi alcançada uma elevada acurácia preditiva (99%), mas então, qual seria a melhor abordagem para o problema?

A resposta para esta pergunta é complexa. 

Primeiro vamos fazer um comparativo entre as técnicas:

Árvores de Decisão 
--- 
* Definição de modelo simplificada, as técnicas de auto-analise das árvores de decisão permitem um modelo simples de se definir 
* Elevado custo de memória, uma arvore de decisão pode chegar a ter uma estrutura muito grande para um dado problema 
* Fácil compreensão e análise 
---
RNA
---
* Definição complexa, devido as inúmeras arquiteturas possíveis, é dificil decidir qual será a mais adequada para a solução do problema
* Um problema que seria necessário uma grande árvore, é resolvido por poucos neurônios
*  Modelo de dificil compreensão e complexo de ser análisado.
---

Apesar das diferenças entre ambos os modelos é notório que ambos alcançaram o objetivo de resolver o problema proposto com soluções de qualidades equiparáveis, no entanto devido ao rápido crescimento na estrutura de uma árvore de decisão a mesma é mais recomendada para problemas de menor complexidade, enquanto problemas mais complexos, podem ser resolvidos por RNAs, mesmo que com certa complexidade.
