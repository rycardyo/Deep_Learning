# Definição da técnica utilizada: Redes Neurais


   Em curtas palavras, uma rede neural nada mais é que uma técnica computacional utilizada para resolução de problemas dificeis de serem mapeados com a computação tradicional, algoritmos simples (famosos 'se' e 'entao').

   Uma rede neural busca simular em um modelo computacional algo que se assemelhe com o funcionamento do cérebro humano, ou seja, um aglomerado de neuronios densamente conectados entre si. 

  Para converter o conceito de neurônio para a computação, é então implementada uma **'arquitetura'** de neurônio a qual é composto por entradas, **pesos na entradas, um processamento (feito 'dentro' do neuronio) e uma saída**, conforme a imagem que segue abaixo:







```
from IPython.display import Image, display

display(Image(filename='neuronio_artificial.jpg'))
```

- Onde x1, x2, ..., xn são os inputs e w1, w2, ..., wn são os respectivos pesos de cada input.

- O $\sum$ é uma combinação linear desses inputs ponderada pelos pesos e ela gera o output do node e é dada por:

$$o = \sum_{i=0}^{n}x_iw_i$$

- Onde o é output desse Node. Depois disso ele é avaliado por uma função de perda. Para problemas de classificação binária é muito comum o uso da função 0-1 loss:

$$l(o) = \left\{\begin{matrix} 1 & se & o \geq 0 \\ 0 & se & o < 0 \end{matrix}\right.$$

Apesar de parecer complexa, a técnica é muito eficiente em problemas de IA, como o nosso problema é de classificar uma movimentação em legitma ou fraudulenta, seria um problema de classificação o qual é feito com bastante competencia pelo modelo proposto acima

## Arquitetura da Rede Neural

Na definição da arquitetura, iremos definir a estrutura da nossa rede, em outras palavras, quantos neuronios iremos utilizar, quantas camadas e quantos neuronios por camada será utilizado.

Existem diversas formas de se definir a arquitetura da rede sendo algumas arquiteturas mais eficientes para dados problemas que outras. Por exemplo, em uma arquitetura de rede neural ideal (que imita 100% de um cerebro humano) teriamos milhoes, talvez bilhoes de neuronios com muitas conexoes entre si. No entanto para o computador executar uma rede tao densa, há um custo computacional que torna inviável a implementação de uma rede tão complexa, por essa razão o ideal é buscar uma rede que seja o mais simples possível ao mesmo tempo que, seja capaz de atender a proposta de solução do problema dado. 


* Função de Ativação:
 - Devido a variavel de interesse ser booleana, a função de ativação pode ser do tipo limear (0 ou 1)

* Topologia da rede: 
 - Diz respeito ao número de camadas, número de neurônios por camada, o padrão de conexões e se existem conexões de retropropagação. A escolha deste passo é mais complexa pois não se trata a primeira vista, de uma avaliação intuitiva, no entanto será aplicada uma abordagem simplificada na definição da arquitetura. Ou seja, inicializaremos a rede com a seguinte arquitetura:

 ### Definição da arquitetura:
 * Será constituída por 3 camadas
 * Cada camada possuíra Total de atributos/2 neuronios ou seja 5 neurônios, com excessão da ultima que possuíra somente 1 única neuronio.
 * Caso a arquitetura se mostre ineficaz, a mesma será reajustada

## Algoritmo/estratégia de treinamento da rede

Para treinamento da RNA será utilizado o algoritmo perceptron devido aos seguintes motivos:
* Utiliza uma função de ativação do tipo limiar
* Possuí boa acurácia em problemas de classificação
* Relativamente simples
* Elimina o risco de ótimos locais, como ocorrem com o algoritmo back-propagation, e, também evita o problema do elevado tempo de convergencia.
